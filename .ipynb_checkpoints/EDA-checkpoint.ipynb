# Importing the neccessary Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the Dataset

```python
data = pd.read_csv("diabetes.csv")
```

```python
data
```

# Display the first few rows of the dataset


```python
print("First few rows of the dataset:")
print(data.head())
```

# Summary statistics

```python
print("\nSummary statistics:")
print(data.describe())
```

```python
data.info()
```

# Checking the datatypes of each column

```python
print("\nData types:")
print(data.dtypes)
```

# Fasting Urin Patients Percentage

```python
fasting_urin_patients_percentage = (data['FU'].value_counts(normalize=True) * 100)['YES']
print(f"The percentage of patients who have a Fasting Urin: {fasting_urin_patients_percentage:.2f}%")
```

# Fasting Sugar Patients Percentage

```python
fasting_sugar_patients_percentage = (data['Diabetes'].value_counts(normalize=True) * 100)['YES']
print(f"The percentage of patients who have Fasting Sugar: {fasting_sugar_patients_percentage:.2f}%")
```

# Pie Chart for Diabetes Status

```python
diabetes_counts = data['Diabetes'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(diabetes_counts, labels=diabetes_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Diabetes Status')
plt.axis('equal')
plt.show()
```

# Pie Chart for Follow-up Status

```python
fu_counts = data['FU'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(fu_counts, labels=fu_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Follow-up Status')
plt.axis('equal')
plt.show()
```

# Histogram of FS (Fasting Sugar)

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='FS', bins=10)
plt.title('Distribution of Fasting Sugar')
plt.xlabel('Fasting Sugar')
plt.ylabel('Frequency')
plt.show()
```

# Countplot of FU (Family History of Diabetes)

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='FU')
plt.title('Family History of Diabetes')
plt.xlabel('Family History')
plt.ylabel('Count')
plt.show()
```

# Boxplot of FS by Diabetes

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=data, x='Diabetes', y='FS')
plt.title('Fasting Sugar by Diabetes')
plt.xlabel('Diabetes')
plt.ylabel('Fasting Sugar')
plt.show()
```

# Pairplot

```python
sns.pairplot(data)
plt.show()
```

# Barplot of Average FS by Diabetes:

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=data, x='Diabetes', y='FS')
plt.title('Average Fasting Sugar by Diabetes')
plt.xlabel('Diabetes')
plt.ylabel('Average Fasting Sugar')
plt.show()
```

# Boxplot of FS by FU and Diabetes

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=data, x='FU', y='FS', hue='Diabetes')
plt.title('Fasting Sugar by Family History of Diabetes and Diabetes')
plt.xlabel('Family History')
plt.ylabel('Fasting Sugar')
plt.legend(title='Diabetes')
plt.show()
```

# PairGrid with Scatterplot and Kernel Density Estimate (KDE):

```python
g = sns.PairGrid(data)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
plt.show()
```

# Categorical Scatterplot with Swarmplot

```python
sns.catplot(data=data, x='Diabetes', y='FS', hue='FU', kind='swarm')
plt.show()
```
